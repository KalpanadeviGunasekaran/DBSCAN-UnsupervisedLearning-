```python
import pandas as pd
import numpy as pn
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
df
```

```python
features=['Annual Income (k$)','Spending Score (1-100)']
X=df[features]
```

```python
X_standarized=StandardScaler().fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(X_standarized)
clusters
```

```python
centroids=pd.DataFrame(X_standarized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'],c=clusters,cmap='viridis',edgecolors='k')
plt.title('DBSCAN Clusters of Mall Customer Data')
plt.xlabel('Annual Income (k$)')
plt.ylabel("Spending Score (1-100)")
plt.show()
```
